### plotly wont show up, gotta show a png file

```python
import plotly.express as px

iris_df = px.data.iris()
print(iris_df)

print(sns.load_dataset("iris")) # same thing
# print statements will show up in github
```

```python

fig = px.scatter_3d(iris_df,
           "sepal_length",
           "sepal_width",
           "petal_length",
           "species",
            opacity = 0.5
           )

fig

```

### Here is the plot if you can not see it!



```python
px.scatter_3d(iris_df,
           x= "sepal_length",
           y= "sepal_width",
           z = "petal_length",
           color = "species",
           symbol = "species",
          opacity = 0.5
           )
```
